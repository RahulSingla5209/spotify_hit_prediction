```python
sixties_df  = pd.read_csv("dataset-of-60s.csv")
seventies_df = pd.read_csv("dataset-of-70s.csv")
eighties_df = pd.read_csv("dataset-of-80s.csv")
nineties_df = pd.read_csv("dataset-of-90s.csv")
twothousands_df = pd.read_csv("dataset-of-00s.csv")
tens_df = pd.read_csv("dataset-of-10s.csv")
pd.set_option("display.max_rows", 500)
pd.set_option("display.max_columns", 500)


sixties_df['decade'] = "60s"
seventies_df['decade'] = "70s"
eighties_df['decade'] = "80s"
nineties_df['decade'] = "90s"
twothousands_df['decade'] = "00s"
tens_df["decade"] = '10s'
frames = [sixties_df,seventies_df,eighties_df,nineties_df,twothousands_df,tens_df]
df = pd.concat(frames)
df.columns
```

```python
ct_key = pd.crosstab(df["key"],df["decade"])

tens_percentiles = ct_key.sort_values(by = "10s",ascending = False)["10s"]/ct_key["10s"].sum()
sixties_percentiles = ct_key.sort_values(by = "60s",ascending = False)["60s"]/ct_key["60s"].sum()

seventies_percentiles = ct_key.sort_values(by = "70s",ascending = False)["70s"]/ct_key["70s"].sum()

eighties_percentiles = ct_key.sort_values(by = "80s",ascending = False)["80s"]/ct_key["80s"].sum()

nineties_percentiles = ct_key.sort_values(by = "90s",ascending = False)["90s"]/ct_key["90s"].sum()

twothousands_percentiles = ct_key.sort_values(by = "00s", ascending = False)["00s"]/ct_key["10s"].sum()

```
